```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
df.head(10)
```

```python
d = df.groupby(['continent'])['total_cases'].agg(np.max)
d4 = df.groupby(['continent'])['new_cases'].agg(np.max)
```

```python
d2 = df.groupby(['continent'])['total_deaths'].agg(np.max)
d3 = df.groupby(['continent'])['new_deaths'].agg(np.max)
```

```python
fix,ax = plt.subplots()
ax.plot(d, color = 'r', marker = 'o')
ax.plot(d4 , color = 'y',marker = 'o')
ax.legend(['total_cases','new_cases'])
```

```python
fix2,ax2 = plt.subplots()
ax2.plot(d2, color = 'b', marker = 'o')
ax2.plot(d3 , color = 'g',marker = 'o')
ax2.legend(['total_deaths','new_deaths'])
```
